```python
# 4.Write a python program to create two threads.Threads one must print the list of squares and thread two must print the list of cubes.

# Ans-
#       
import threading
    
def print_squares():
        for i in range(1, 6):
            print(f"Square of {i} is {i**2}")
            
def print_cubes():
    for i in range(1, 6):
        print(f"Cube of {i} is {i**3}")
            
if __name__ == "__main__":
    t1 = threading.Thread(target=print_squares)
    t2 = threading.Thread(target=print_cubes)
        
    t1.start()
    t2.start()
        
    t1.join()
    t2.join()
        
    print("Done")

        
# output:
#     square of 1 is 1
#     cube of 1 is 1
#     square of 2 is 4
#     cube of 2 is 8
#     square of 3 is 9
#     cube of 3 is 27
#     square of 4 is 16
#     cube of 4 is 64
#     square of 5 is 25
#     cube of 5 is 125
#     Done
    
```
